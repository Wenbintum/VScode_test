```python
import pandas as pd
import numpy as np
import xlrd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
datas=pd.read_excel('data.xlsx',sheet_name='refining', na_values='n/a')   ###read first sheet1
```

```python
datas.head()
```

```python
lrbt = [0.135,0.7,0.25,0.78]
rcParams['figure.subplot.left'] = lrbt[0]   # the left side of the subplots of the figure
rcParams['figure.subplot.right'] = lrbt[1]  # the right side of the subplots of the figure
rcParams['figure.subplot.bottom'] = lrbt[2] # the bottom of the subplots of the figure
rcParams['figure.subplot.top'] = lrbt[3]    # the top of the subplots of the figure

fig, ax = plt.subplots(1,1)
datas.plot(ax=ax,color=['red','blue','green','orange','black'], marker='o',markersize=5, x='xnumber',y=['train-0r','train-1r','train-2r','train-3r','Sca-train'])
datas.plot(linestyle='--',color=['red','blue','green','orange','black'], marker='*', markersize=8, markerfacecolor='none',ax=ax,x='xnumber',y=['vali-0r','vali-1r','vali-2r','vali-3r','Sca-vali'])
#datas.plot.scatter(datas['all-2r'])fi, ax = plt.subplots()

ax.set_xlabel("Dimension",fontsize=12)
ax.set_ylabel("RMSE (eV)",fontsize=12)
plt.legend([ "$\Phi$0 Training",
             "$\Phi$1 Training",
             "$\Phi$2 Training",
             "$\Phi$3 Training",
             "Scal. Training",
             "$\Phi$0 Validation",
             "$\Phi$1 Validation",
             "$\Phi$2 Validation",
             "$\Phi$3 Validation",
             "Scal. Validation"],loc='center right',bbox_to_anchor=(1.5, 0.5))
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlim(0.5, 8.5)
plt.ylim(0.05, 0.35)
plt.title('Training and Validation of OH* and OOH*',fontsize=12)
#plt.plot(datas.index, datas, 'o', color='green')
plt.savefig('compare-scaling-2adsorbate.png',dpi=300)
plt.show()
```
